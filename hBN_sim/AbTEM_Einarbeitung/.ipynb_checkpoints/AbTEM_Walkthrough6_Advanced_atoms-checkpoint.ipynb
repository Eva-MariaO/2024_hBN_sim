```python
import ase
import matplotlib.pyplot as plt
import numpy as np
import abtem

from ase.cluster.cubic import FaceCenteredCubic
```

# Advanced Atomis

### Setup

```python
#creating nanoparticle by using the ase.cluster module 
#therefore we define directions (surfaces named by miller indices) and number of layers in those directions (counting from center)

surfaces1 = [(1,0,0), (1,1,0)]
layers1 = (5,12)
nanoparticle1 = FaceCenteredCubic("Ag", surfaces1, layers1) #silber struktur 

#truncated
surfaces2 = [(1,0,0), (1,1,0), (1,1,0)]
layers2 = (5,7,1)
nanoparticle2 = FaceCenteredCubic("Ag", surfaces2, layers2) #silber struktur 

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(11,5))
abtem.show_atoms(nanoparticle1, ax=ax1, scale=0.8)
abtem.show_atoms(nanoparticle1, ax=ax2, scale=0.8, plane="xz", title="xz full")

abtem.show_atoms(nanoparticle2, ax=ax3, scale=0.8)
abtem.show_atoms(nanoparticle2, ax=ax4, scale=0.8, plane="xz", title="xy trunc")

plt.tight_layout()
```

```python
#walkthrough example
surfaces = [(1, 0, 0), (1, 1, 0), (1, 1, 1)]
layers = [6, 9, 5]
lc = 4.08

nanoparticle = FaceCenteredCubic("Au", surfaces, layers, latticeconstant=lc)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9,4))
abtem.show_atoms(nanoparticle, ax=ax1, scale=0.9)
abtem.show_atoms(nanoparticle, ax=ax2, scale=0.9, plane="xz")
```

```python
centered_nanoparticle = nanoparticle.copy()
y_angle = np.degrees(np.arctan(1/np.sqrt(2)))
centered_nanoparticle.rotate(45, 'x')
centered_nanoparticle.rotate(y_angle, 'y')
centered_nanoparticle.center(vacuum=2)

abtem.show_atoms(centered_nanoparticle, scale=1)
```

### Manipulation

```python
#rotation needed because abtem always assumes electrons propagating from negative to positive z direction
#rotate particle into the (111) zone axis -> zuerst um 45 um 

rotated_nanoparticle = nanoparticle1.copy()
rotated_nanoparticle1 = nanoparticle1.copy()

rotated_nanoparticle.rotate(45, "x")
rotated_nanoparticle1.rotate(-45, "z")

y_angle = np.degrees(np.arctan(1 / np.sqrt(2)))

end_particle = rotated_nanoparticle.copy()
end_particle1 = rotated_nanoparticle1.copy()

end_particle.rotate(y_angle, "y")
end_particle1.rotate((90-y_angle), "x")


fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2, 2, figsize=(11,4))
abtem.show_atoms(rotated_nanoparticle, scale=0.9, ax=ax1, title="rotation around x-axis by 45")
abtem.show_atoms(end_particle, scale=0.9, ax=ax2, title="rotation around y-axis by ~36")
abtem.show_atoms(rotated_nanoparticle1, scale=0.9, ax=ax3, title="rotation around z-axis by -45")
abtem.show_atoms(end_particle1, scale=0.9, ax=ax4, title="rotation around y-axis by 90-36")


plt.tight_layout()
```

### Mix & Match

```python
repeated_srtio3 = srtio3_110 * (2, 6, 4)

#sto_lto: strontiumtitanat & lanthan titanat
sto_lto_interface = repeated_srtio3.copy()

# select atoms with atomic number 38 = strontium
mask_sr = sto_lto_interface.numbers == 38

# select atoms below the center
mask_below = sto_lto_interface.positions[:, 1] < sto_lto_interface.cell[1, 1] / 2

# combine selection
mask_combined = mask_sr * mask_below

# assign new atomic numbers to selection
sto_lto_interface.numbers[mask_combined] = 57
abtem.show_atoms(sto_lto_interface, legend=True);
```

```python
fcc = ase.cluster.cubic.FaceCenteredCubic('Cu', ((0,0,1), (1,1,0)), (5,5))

specific_position = fcc.positions[-1]
mask = np.all((fcc.positions == specific_position), axis=1)#alle 3 koordinaten mssen gleich sein -> np.all mit axis=1 (hier zeilen, axis=0 wren spalten)

mask.shape
fcc_manipulated = fcc.copy()
fcc_manipulated.numbers[mask] = 6
abtem.show_atoms(fcc_manipulated, plane="xy")

#klappt noch nicht, evtl weiter spter
```

### building stuff

```python
substrate = ase.build.bulk("C", cubic=True)

# repeat diamond structure
substrate = substrate * (10, 10, 10)

# displace atoms with a standard deviation of 50 % of the bond length
bl = 1.54  # Bond length
rng = np.random.default_rng(seed=10)#default_rng is random generator
substrate.positions[:] += rng.normal(size=(len(substrate), 3)) * 0.5 * bl

# wrap the atoms displaced outside the cell back into the cell
sub_wrap = substrate.copy()
sub_wrap.wrap() #-> quetscht Auenatome in Zellrahmen, Anzahl bleibt gleich.
#substrate.wrap()

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(9,4))
abtem.show_atoms(substrate, ax=ax1, title='substrate')
abtem.show_atoms(sub_wrap, ax=ax2, title='+ wrap')
```

```python
#use nanoparticle from walkthrough example -> cut particles below 5 angstrom on z-axis
mask_cut = centered_nanoparticle.positions[:, 2] < 5
nanoparticle_cut = centered_nanoparticle.copy()
print(len(nanoparticle_cut))

#del braucht indizes
del nanoparticle_cut[np.where(mask_cut)]#np.where(mask) gibt mit array mit indizes aus, wo Bedingung wahr
print(len(nanoparticle_cut))

# center nanoparticle relative to substrate
nanoparticle_cut.set_cell(substrate.cell)
nanoparticle_cut.center()

# shift nanoparticle in the z-direction
nanoparticle_cut.positions[:, 2] += 27
```

```python
#combi
supported_nanoparticle = substrate + nanoparticle_cut
supported_nanoparticle.center(axis=2, vacuum=5)

#supported_nanoparticle.rotate(180, 'x')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 5))
abtem.show_atoms(supported_nanoparticle, ax=ax1, title="Supported nanoparticle beam view",merge=0.5, scale=1)
abtem.show_atoms(supported_nanoparticle, ax=ax2, plane="xz", scale=1, title="Supported nanoparticle side view", merge=0.5,)
fig.tight_layout();
```

### Orthogonal and periodic supercells

```python
#tackle problem of orthogonalizing cells
graphene = ase.build.graphene(vacuum=2)

graphene_orth, trafo = abtem.orthogonalize_cell(graphene, max_repetitions=10, return_transform=True)

#
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))
abtem.show_atoms(graphene, ax=ax1, title='unit cell')
abtem.show_atoms(graphene_orth, ax=ax2, title='orthogonalized cell')
```

```python
from abtem.atoms import pretty_print_transform
pretty_print_transform(trafo)
#we can see that no additional strain was needed  (orthogonalizing hex cell worked by doubling cell size)
```

```python
#other example -> rotated gold structure
desired_rotation = 30
gold = ase.build.bulk("Au", cubic=True)
gold.rotate(desired_rotation, "y", rotate_cell=True)

gold_orth, trafo_gold = abtem.orthogonalize_cell(gold, max_repetitions=10, return_transform=True)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 10))

abtem.show_atoms(gold, ax=ax1, scale=0.5, title="Rotated gold beam view")
abtem.show_atoms(gold, ax=ax2, plane="xz", scale=0.5, title="Rotated gold side view");
abtem.show_atoms(gold_orth, ax=ax3, scale=0.5, title="Rotated gold beam view")
abtem.show_atoms(gold_orth, ax=ax4, plane="xz", scale=0.5, title="Rotated gold side view");
```
